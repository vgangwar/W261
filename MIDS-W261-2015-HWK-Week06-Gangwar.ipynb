####Answer
**For unconstrained univariate optimization what are the first order Necessary Conditions for Optimality (FOC)?**  
>For unconstrained univariate optimization the first order Necessary Conditions for Optimality (FOC) is that the first order derivative of the objective function evaluates to zero at candidate extremum  
$f'(x=x^*)=0$

** What are the second order optimality conditions (SOC)?**  
>The second order optimality conditions (SOC) is that for the points where the 1st derivative evaluates to zero, if the 2nd derivative evaluates to less than zero, then that point is the local maximum. Conversely, if the 2nd order derivative evaluates to greater than 0 then that point is the local minimum  
if $f''(x=x*) < 0$, then maximum  
if $f''(x=x*) > 0$, then minimum  
if $f''(x=x*) = 0$, then inconclusive

**For unconstrained multi-variate optimization what are the first order Necessary Conditions for Optimality (FOC)?**  
>The gradient of the function, which is a vector whose components are partial derivatives of the objective function in each dimension, evaluates to zero at the candidate extremum  
$\nabla f(X=x*) = (\frac{\partial f}{\partial x_1}, \frac{\partial f}{\partial x_2}, ..., \frac{\partial f}{\partial x_n}) = (0, 0, ..., 0)$

**What are the second order optimality conditions (SOC)?**  
>If the Hessian is positive definite at $x$, then the objective function attains a local minimum at $x$. If the Hessian is negative definite at $x$, then the objective function attains a local maximum at $x$

**What is the Hessian matrix in this context?**  
>Hessian matrix is a square matrix of second-order partial derivatives of a multi-dimensional objective function. In this is context it is used in the Newton-Raphson method to find roots of a gradient function. The determinant of the Hessian matrix (called the Hessian) is used to determine whether a critical point is a maxima or minima

**What makes an optimization problem convex?**  
> An optimization problem is convex if:
- If the domain is a convex set
- The objective functioin is convex
- Any local minimum is a global minimum
- The function passes the chord test i.e. a line segment drawn from any point $(x_1, f(x_1))$ to another point $(x_2, f(x_2))$ lies on or above the graph of $f$
- The inequality constracints are convex
- The equality contraints are affine

**What are the first order Necessary Conditions for Optimality in convex optimization?**  
>The function is globally above the tangent at any point $(x, f(x))$

** What are the second order optimality conditions for convex optimization?**  
>The function is flat or curved upwards in every direction

**Are both necessary to determine the maximum or minimum of candidate optimal solutions?**  
>Only FoC is needed. SoC is not needed

#HW 6.4
The learning objective function for weighted ordinary least squares (WOLS) (aka weight linear regression) is defined as follows:

$0.5* sumOverTrainingExample i (weight_i * (W * X_i - y_i)^2)$

Where training set consists of input variables X ( in vector form) and a target variable y, and W is the vector of coefficients for the linear regression model.

Derive the gradient for this weighted OLS by hand; showing each step and also explaining each step.

####Answer
Objective Function = $ J(W) = 0.5 * \sum_{i=1}^m (weight_i(W_jX_i - y_i)^2)$  
Subscript $i$ denotes $m$ training examples  
$W$ are the model parameters with a total of $j$ parameters  
To derive the gradient, calculate the partial derivative of the objective function with respect to W  
The gradient of the objective function is a gradient vector of partial derivatives with respect to the model parameters  
  
$\nabla J(W) = \frac{\partial (0.5 * \sum_{i=1}^m (weight_i(W_jX_i - y_i)^2))}{\partial W_j}$  
  
0.5 and Summation comes out the derivative  
$= 0.5 * \sum_{i=1}^m\frac{\partial (weight_i(W_jX_i - y_i)^2)}{\partial W_j}$  

Applying chain rule  
the common term 2 comes out of the summation  
$= 0.5 * 2 * \sum_{i=1}^m weight_i(W_jX_i - y_i)\frac{\partial(W_jX_i - y_i)}{\partial W_j}$  
  
The index $j$ denotes the number of vector components of the gradient vector  
$= \sum_{i=1}^m weight_i(W_jX_i - y_i)X_j$

```python
%%writefile MrJobBatchGDUpdate_LinearRegression.py
from mrjob.job import MRJob
from mrjob.step import MRStep

# This MrJob calculates the gradient of the entire training set 
#     Mapper: calculate partial gradient for each example  
#     
class MrJobBatchGDUpdate_LinearRegression(MRJob):
    # run before the mapper processes any input
    def read_weightsfile(self):
        # Read weights file
        with open('weights.txt', 'r') as f:
            self.weights = [float(v) for v in f.readline().split(',')]
        # Initialze gradient for this iteration
        self.partial_Gradient = [0]*len(self.weights)
        self.partial_count = 0
    
    # Calculate partial gradient for each example 
    def partial_gradient(self, _, line):
        D = (map(float,line.split(',')))
        # y_hat is the predicted value given current weights
        y_hat = self.weights[0]+self.weights[1]*D[1]
        # Update parial gradient vector with gradient form current example
        #self.partial_Gradient =  [self.partial_Gradient[0]+ D[0]-y_hat, self.partial_Gradient[1]+(D[0]-y_hat)*D[1]]
        # Adding weights
        # -----
        self.partial_Gradient[0] = self.partial_Gradient[0] + (D[0]-y_hat)/abs(D[1])
        self.partial_Gradient[1] = self.partial_Gradient[1] + (D[0]-y_hat)*D[1]/abs(D[1])
        # -----
        self.partial_count = self.partial_count + 1
        #yield None, (D[0]-y_hat,(D[0]-y_hat)*D[1],1)
    
    # Finally emit in-memory partial gradient and partial count
    def partial_gradient_emit(self):
        yield None, (self.partial_Gradient,self.partial_count)
        
    # Accumulate partial gradient from mapper and emit total gradient 
    # Output: key = None, Value = gradient vector
    def gradient_accumulater(self, _, partial_Gradient_Record): 
        total_gradient = [0]*2
        total_count = 0
        for partial_Gradient,partial_count in partial_Gradient_Record:
            total_count = total_count + partial_count
            total_gradient[0] = total_gradient[0] + partial_Gradient[0]
            total_gradient[1] = total_gradient[1] + partial_Gradient[1]
        yield None, [v/total_count for v in total_gradient]
    
    def steps(self):
        return [MRStep(mapper_init=self.read_weightsfile,
                       mapper=self.partial_gradient,
                       mapper_final=self.partial_gradient_emit,
                       reducer=self.gradient_accumulater)] 
    
if __name__ == '__main__':
    MrJobBatchGDUpdate_LinearRegression.run()
```

####MapReduce job for generating 1% sample
####Code Description
- Mapper generates a random floating point number between 0 and 1.0 and converts it to str and removes the decimal point
- Mapper emits the random number (in str format) as key and the data point as value
- Framework sorts keys and presents it to the single reduce
- The reducer yields the 1st 1% of the training data

```python
%%writefile sampler.py
from mrjob.job import MRJob
from mrjob.step import MRStep
import random
   
class sampler(MRJob):
    def mapper(self, _, line):
        r = random.random()
        r = str(r).replace('.', '')
        yield r, line
    
    def reducer_init(self):
        self.threshold = 10000
        self.counter = 0
        
    def reducer(self, key, values):
        for value in values:
            if self.counter < self.threshold:
                yield key, value
                self.counter += 1
    
    def steps(self):
        return [MRStep(mapper       = self.mapper,
                       reducer_init = self.reducer_init,
                       reducer      = self.reducer)] 
    
if __name__ == '__main__':
    sampler.run()
```

```python
import statsmodels
import numpy as np
import statsmodels.api as sm

Y = list()
X = list()

filename = 'sample.txt'
with open(filename, 'r') as filehndl:
    for line in filehndl:
        line = line.strip().replace("'", "").replace('u', '').replace('(', '').replace(')', '')
        [_, y, x] = line.split(',')
        y = float(y)
        x = float(x)
        Y.append(y)
        X.append(x)
    
Y = np.array(Y).reshape(10000, 1)
X = np.array(X).reshape(10000, 1)
X = sm.add_constant(X)
weights = 1 / np.absolute(X)


mod_wls = sm.WLS(Y, X, w=weights)
res_wls = mod_wls.fit()
print res_wls.summary()
```
